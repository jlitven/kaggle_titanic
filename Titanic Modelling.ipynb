# Titanic Modelling

The challenge: Find a model to predict which passengers on the titanic survived.

In this notebook, we will build models to perform the predictions and submit them to kaggle!


The kaggle challenge can be found here:
https://www.kaggle.com/c/titanic-gettingStarted

The exploratory analysis and modelling use the training data [train.csv](https://www.kaggle.com/c/titanic/download/train.csv)  and test data [test.csv](https://www.kaggle.com/c/titanic/download/test.csv).

## Cleaning
As a preliminary step we need to clean the data by fixing missing values. Since missing values can appear in the training and test data, it is useful to treat them as a single data frame.

```python
# Required libraries
import pandas as pd
import numpy as np

# Import the data
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')

# Glue training and test set together
df = pd.concat([df_train, df_test], axis=0)
left_columns = ['Survived', 'PassengerId']
columns_ordered = left_columns + [c for c in df.columns if c not in left_columns]
df = df[columns_ordered]

df.head()
```

A brief description of the columns:
* Survived - Whether the passenger survived or perished
* PassengerId - Unique index
* Age - Passenger's age
* Cabin - The cabin the passenger stayed in
* Embarked - Which city the passenger departed from
* Fare - How much the passenger paid for the ticket
* Name - The passenger's name
* Parch - The number of parents are children also on board
* Pclass - The class of the passenger
* Sex - The sex of the passenger
* SibSp - How many siblings or spouses also on board
* Ticket - The ticket number

Let's start by dropping the features we won't be using in the analysis.

```python
df = df.drop(['Cabin', 'Ticket', 'Embarked'], axis=1)
```

Which remaining features have null values?

```python
df.isnull().any()
```

We expect Survived to have null values because of the test set. For 'Age' and 'Fare', we will use the median values of the passenger's class.

```python
df_class = df.groupby('Pclass').median()
df_class.head()
```

```python
# Fill in missing values for Age
null_ages = df['Age'].isnull() == True
df['Age'].loc[null_ages] = df['Pclass'].loc[null_ages].apply(lambda pclass: df_class.ix[pclass]['Age'])

# Fill in missing values for Fare
null_fares = df['Fare'].isnull() == True
df['Fare'].loc[null_fares] = df['Pclass'].loc[null_fares].apply(lambda pclass: df_class.ix[pclass]['Fare'])
```

## Feature Engineering

Let's create two new features: FamilySize, which counts the total number of family members on board, and FamilyID, which groups family members together.

```python
# Create family size feature
df['FamilySize'] = df['SibSp'] + df['Parch']
df = df.drop(['SibSp', 'Parch'], axis=1)

# To group family members, identify them by their surname and family size
surnames = df['Name'].apply(lambda name: name.split(',')[0])
family_ids = df['FamilySize'].astype(str) + surnames
family_ids_count = family_ids.value_counts()
df['FamilyID'] = family_ids.apply(lambda id: id if family_ids_count[id] > 3 else 'small')
df = df.drop(['Name'], axis=1)
```

We need to create dummy variables for our categorical features: Sex, Pclass, and FamilyID.

```python
# Create dummy variables by concatenating dataframe with dummy columns
for col in ['Sex', 'Pclass', 'FamilyID']:
    df = pd.concat([df, pd.get_dummies(df[col], prefix=col, drop_first=True)], axis=1)
    
# Now that we've created dummy columns we can drop categorical variables
df = df.drop(['Sex', 'Pclass', 'FamilyID'], axis=1)

# Split back into training and testing
n_train = df_train.shape[0]
df_test = df[n_train:]
df = df[:n_train]

X = df.ix[:, 2:]
X_test = df_test.ix[:, 2:]
y = df['Survived']
```

Now we are ready to model!

## Modelling
### Model: Women and Children First
Let's start with a simple model using the popular notion that mainly women and children survived. We will predict that all women and children survived, while all men perished.

```python
def women_and_children_model(passenger):
    age, male = passenger
    if age < 16 or male == 0:
        return 1
    else:
        return 0
```

```python
# Apply the model to the training data
from sklearn import metrics

y = df.Survived
y_prediction = X[['Age', 'Sex_male']].apply(women_and_children_model, axis=1)
print 'Accuracy on training data :{}'.format(metrics.accuracy_score(y, y_prediction))
```

```python
# Apply the model to the test data and submit to kaggle
pred_women_and_children = df_test[['Age', 'Sex_male']].apply(women_and_children_model, axis=1)
df_test['Survived'] = pred_women_and_children.astype(int)

df_result = df_test[['Survived', 'PassengerId']]
df_result.to_csv('models/women_and_children_model.csv', index=False)
```

```python
# Kaggle Result: Your submission scored 0.76077
```

The women and children model resulted in a score of 0.76077. Let's see if we can improve our model by using a machine learning technique known as Decision Trees.

### Model: Decision Tree

Decision Trees determine a prediction of a passenger (or any data in general) by asking questions about the passenger, for example, is the passenger a male or female? Questions are asked, starting at a 'root' question, until it arrives at a 'leaf' of the tree. Each leaf will base its prediction on the training data which have answered the questions in the same way. If the majority of training data in a given leaf has survived, then that leaf is marked 'survived'. Below is an example tree created from the training data.

```python
from IPython.display import Image
import pydot
from StringIO import StringIO
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz

clf_tree = DecisionTreeClassifier(max_depth=2)
clf_tree = clf_tree.fit(X, y)
dot_data = StringIO()  
export_graphviz(clf_tree, 
                out_file=dot_data,
                filled=True,
                feature_names=X.columns,
                class_names=['Perished', 'Survived'])

graph = pydot.graph_from_dot_data(dot_data.getvalue())[0] 
Image(graph.create_png())  
```

Let's create a full decision tree and see how it performs.

```python
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier(max_features=15, max_depth=7)
clf_tree = clf_tree.fit(X, y)

y_prediction = clf_tree.predict(X)
print 'Accuracy on training data: {}'.format(metrics.accuracy_score(y, y_prediction))
```

Wow, it scored 87%! That's dubiously high, let's run it on the test data.

```python
pred_decision_tree = clf_tree.predict(X_test)
df_test['Survived'] = pred_decision_tree.astype(int)

df_result = df_test[['Survived', 'PassengerId']]
df_result.to_csv('models/decision_tree.csv', index=False)
```

```python
# Kaggle Result: Your submission scored 0.77512
```

So the decision trees is better than our naive model by about 1%. We can also view the importance of each feature from the decision tree. 

```python
feature_importance = zip(X.columns, list(clf_tree.feature_importances_))
feature_importance = sorted(feature_importance, key=lambda (f, s): s, reverse=True)

print '\n'.join(['{}: {}'.format(f,i) for f, i in feature_importance])
```

While the bias  of decision trees is low, the variance is high because they often overfit to the training data. To ameliorate this we turn to the approach of Random Forests.

### Model: Random Forest
A random forest is an ensemble method which creates many decision trees over subsamples of the data. By leveraging the use of many trees the estimator avoids the problem of overfitting to the training data. Subsamples are generated by selecting the original data with replacement. Let's create a random forest with 1000 decision trees.

```python
from sklearn.ensemble import RandomForestClassifier

clf_random_forest = RandomForestClassifier(n_estimators=1000, max_features=15, max_depth=7) # tuned
clf_random_forest = clf_random_forest.fit(X, y)

y_prediction = clf_random_forest.predict(X)
print 'Accuracy on training data: {}'.format(metrics.accuracy_score(y, y_prediction))
```

Again the accuracy on the training data is quite high. How does it perform on the test set?

```python
pred_random_forest = clf_random_forest.predict(X_test)
df_test['Survived'] = pred_random_forest.astype(int)

df_result = df_test[['Survived', 'PassengerId']]
df_result.to_csv('models/random_forest.csv', index=False)
```

```python
# Kaggle Result: Your submission scored 0.78469
```

We see that the random forest improves on the decision tree by about 1%. This may not seem like much, but it raised the position on the scoreboard from 2612 to 1590! Clearly improving the model's accuracy is quite a challenge.
